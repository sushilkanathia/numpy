```python
RANDOM NUMBER GENERATION.

import numpy as np   # Here we can import numerical library (numpy) as np.
```

```python
y = np.random.normal(size = (1,5))[0] #here we can assign a value to y which print any random values according to the size
```

```python
print(y)  #Here we print the value of y
```

```python
z = np.random.randint(low = 2,high = 50,size = 4) #here we can assign a valur to the z and it gernerates some random values according to the size here.
```

```python
print(z)  #Here print the value of z
```

```python
np.random.permutation(y) #Return a new ordering of elements of z
```

```python
np.random.uniform(size=4) #Uniform distribution
```

```python
np.random.normal(size = 4) #Normal distribution
```
